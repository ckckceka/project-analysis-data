```python
import numpy as np
import matplotlib.pyplot as plt 
model = ["Innova","Juke","Pajero","Ertiga","Swift","CRV","Xpander","Brio","Avanza","March","Jazz","Land Cruiser"]
contribution = [10.01,9.11,8.77,8.66,8.55,8.21,8.10,8.10,8.10,7.76,7.54,7.09]
plt.figure(figsize=(12, 6))
plt.bar(model, contribution)
plt.xlabel("Model")
plt.ylabel("Contribution (%)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt 
categories = ['January', 'Febuary', 'March', 'April']
percentage_values = [25.5, 26.4, 24.8, 23.3]

plt.pie(percentage_values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Contribution Of Sales Based On Month")
plt.show()
```
